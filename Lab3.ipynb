# Завдання 1

```python
import numpy as np
import matplotlib.pyplot as plt

h = 1
x = np.arange(-8, 8 + h, h)

y = 2 - np.cos(x)
z = -0.05 * (x**2 + 10 * np.cos(x))

plt.figure(figsize=(10, 6))

plt.plot(x, y, "y*", label=r"$y = 2 - \cos x$")
plt.plot(x, z, "rs", label=r"$z = -0.05(x^2 + 10\cos x)$")

plt.xlabel("x", fontsize=12)
plt.ylabel("y, z", fontsize=12)
plt.title(r"Графіки функцій $y = 2 - \cos x$ та $z = -0.05(x^2 + 10\cos x)$", fontsize=14)

plt.grid(True)
plt.legend()


plt.show()
```

# Завдання 2

```python
import numpy as np
import matplotlib.pyplot as plt

h = 1
x = np.arange(-8, 8 + h, h)

y = 2 - np.cos(x)
z = -0.05 * (x**2 + 10 * np.cos(x))

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

ax1.plot(x, y, "y*", label=r"$y = 2 - \cos x$")
ax1.set_xlabel("x")
ax1.set_ylabel("y(x)")
ax1.set_title(r"Графік функції $y(x) = 2 - \cos x$")
ax1.grid(True)
ax1.legend()
ax1.text(-7.5, 2.3, "Жовта лінія зі зірочками", fontsize=9, color="blue")

ax2.plot(x, z, "rs", label=r"$z = -0.05(x^2 + 10\cos x)$")
ax2.set_xlabel("x")
ax2.set_ylabel("z(x)")
ax2.set_title(r"Графік функції $z(x) = -0.05(x^2 + 10\cos x)$")
ax2.grid(True)
ax2.legend()
ax2.text(-7.5, 0.5, "Червона лінія з квадратами", fontsize=9, color="blue")

fig.suptitle("Графіки функцій y(x) та z(x)", fontsize=14)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

# Завдання 3

```python
import numpy as np
import matplotlib.pyplot as plt

h = np.pi / 100
theta = np.arange(0, 2 * np.pi + h, h)

r = np.cos(3/2 * theta)

plt.figure(figsize=(6, 6))
ax = plt.subplot(polar=True)
ax.plot(theta, r, "m-", linewidth=2, label=r"$r = \cos\left(\frac{3}{2}x\right)$")

ax.set_title("Полярний графік функції r = cos(3/2·x)", va="bottom")
ax.legend(loc="upper right")

plt.show()
```

# Завдання 4

## Частина 1

```python
import numpy as np
import matplotlib.pyplot as plt

h = np.pi / 20
x = np.arange(-2 * np.pi, 2 * np.pi + h, h)
y = np.arange(-2 * np.pi, 2 * np.pi + h, h)

X, Y = np.meshgrid(x, y)

Z = X * Y + np.exp(-(X + Y)) - 3 * X   # xy + 1/e^(x+y) - 3x

plt.figure(figsize=(7, 6))
contour = plt.contour(X, Y, Z, levels=[0], colors="red")
plt.xlabel("x")
plt.ylabel("y")
plt.title(r"Неявний графік: $xy + e^{-(x+y)} - 3x = 0$")
plt.grid(True)
plt.show()
```

## Частина 2

```python
import numpy as np
import matplotlib.pyplot as plt

h = np.pi / 20
t = np.arange(-2 * np.pi, 2 * np.pi + h, h)

x = 2 * t + 3 * t**2
y = -3 * t**2 + 2 * t

plt.figure(figsize=(8, 6))
plt.plot(x, y, "b-", linewidth=2,
         label=r"$x = 2t + 3t^2,\ y = -3t^2 + 2t$")
plt.xlabel("x(t)")
plt.ylabel("y(t)")
plt.title("Параметричний графік на інтервалі [-2π, 2π]")
plt.legend()
plt.grid(True)
plt.axis("equal")
plt.show()
```

```python
import sympy as sp
from sympy.plotting import plot3d

x, y = sp.symbols("x y")
z = x*y + 1/sp.exp(x + y) - 3*x   # z = xy + 1/e^(x+y) - 3x

plot3d(
    z,
    (x, -5, 5),
    (y, -5, 5),
    title="Спосіб 1: 3D поверхня",
    xlabel="x",
    ylabel="y",
    zlabel="z",
)
```

## Частина 3

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)

Z = X*Y + 1/np.exp(X + Y) - 3*X   # xy + 1/e^(x+y) - 3x

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, cmap="viridis")

ax.set_title("Спосіб 3: 3D поверхня (аналог ezmesh)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5.2, 0.2)
y = np.arange(-5, 5.2, 0.2)
X, Y = np.meshgrid(x, y)

Z = X*Y + 1/np.exp(X + Y) - 3*X   # xy + 1/e^(x+y) - 3x

fig, ax = plt.subplots(figsize=(7, 6))

contour = ax.contour(X, Y, Z, levels=30, cmap="jet")
ax.clabel(contour, inline=True, fontsize=8)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Аналог ezcontour для z = xy + 1/e^(x+y) - 3x")

plt.grid(True)
plt.show()
```
