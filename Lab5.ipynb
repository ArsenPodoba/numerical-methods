# Завання 1

```python
import sympy as sp

x = sp.Array([-5.5, -3.5, -1.5, 0.5, 3.5])
y = sp.Array([-6, -3, -1, 3, 6])
```

```python
import sympy as sp

A = sp.Matrix([[val**i for i in range(len(x))] for val in x])

A
```

```python
import sympy as sp

B = sp.Matrix([[v] for v in y])

B
```

```python
import sympy as sp

X = A**-1 * B

X
```

```python
import matplotlib.pyplot as plt
import numpy as np

h = 0.1

x1 = np.arange(-10, 10 + h, h)
y1 = sum(X[i] * x1**i for i in range(len(X)))

plt.figure(figsize=(8, 6))
plt.plot(x, y, "*b", markersize=10, label="Висота")
plt.plot(x1, y1, "r", label="Звичайний многочлен")
plt.title("Графік ")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

# Завдання 2

```python
import numpy as np

X = np.array([14, 16, 18, 22, 26, 30, 32])
Y = np.array([3.6, 7.2, 8.9, 13.3, 19.2, 25.3, 28.6])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(X, Y, "*b", markersize=10, label="Ступінь товщини")
plt.title("Графік розміщення точок (вага біомаси - ступінь товщини)")
plt.xlabel("Вага біомаси")
plt.ylabel("Ступінь товщини")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X1 = X  # лінійна функція

S1 = np.sum(X1)
S2 = np.sum(X1**2)

A = np.array([[S2, S1],
              [S1, len(X1)]])

b1 = np.sum(X1 * Y)
b2 = np.sum(Y)
B = np.array([b1, b2])

K = np.linalg.solve(A, B)
k1, k0 = K  # коефіцієнти апроксиманти

print(f"Формула апроксиманти: y = {k1:.4f} * x + {k0:.4f}")

# Побудова графіка
x_vals = np.linspace(min(X), max(X), 200)
y_vals = k1 * x_vals + k0

plt.scatter(X, Y, color="red", label="Вихідні дані")
plt.plot(x_vals, y_vals, color="blue", label="Лінійна апроксиманта")
plt.grid(True)
plt.title("Лінійна апроксимація")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
