Вхідні дані:

```python
m = 3
n = 7
```

Матриця коефіцієнтів A:

```python
import sympy as sp

A = sp.Matrix([
    [2*m,            5*n,        1 + n,        -5.2],
    [2*m/(2*n + 1),  5.7,       -2.6,          -5.4],
    [-n,             2 + m,      20.5,         -3*n/(m + 1)],
    [n + m,          17.7,      -(m + 1),      n*m]
])

A
```

Вектор вільних членів B

```python
import sympy as sp

B = sp.Matrix([
    4 - 2*m,
    2*n - m,
    -3*m,
    1 + n
])

B
```

Перевірка сумісності


```python
detA = A.det()
rankA = A.rank()
rankAb = A.row_join(B).rank()

print(f"\nВизначник матриці A: {detA}")
print(f"Ранг матриці A: {rankA}")
print(f"Ранг розширеної матриці [A|B]: {rankAb}")

if detA != 0 and rankA == rankAb:
    print("Система сумісна і має єдиний розв’язок.")
elif detA == 0 and rankA == rankAb:
    print("Система сумісна, але має нескінченну кількість розв’язків.")
else:
    print("Система несумісна (розв’язків немає).")
```

Розв’язок за Крамером

```python
if detA != 0:
    xs_cramer = []
    for j in range(A.cols):
        Aj = A.copy()
        Aj[:, j] = B
        xs_cramer.append(Aj.det() / detA)

    print("\nРозв’язок за методом Крамера:")
    for i, x in enumerate(xs_cramer, start=1):
        print(f"x{i} = {sp.simplify(x)}")
```

Розв’язок матричним методом

```python

if detA != 0:
    xs_matrix = A.inv() * B
    print("\nРозв’язок матричним методом:")
    for i, x in enumerate(xs_matrix, start=1):
        print(f"x{i} = {sp.simplify(x)}")
```
