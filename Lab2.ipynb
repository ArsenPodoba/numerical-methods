# Завдання 1

```python
import sympy as sp

x = sp.symbols('x')

# First limit
expr1 = (sp.log(1 + x + x**2) + sp.asin(3*x) - 5*x**3) / (sp.sin(2*x) + sp.tan(x)**2 + (sp.exp(x) - 1)**5)
limit1 = sp.limit(expr1, x, 0)

limit1
```

```python
import sympy as sp

a = sp.symbols('a')

# Second limit
expr2 = (1 - a**x) / (a * x)
limit2 = sp.limit(expr2, x, 0)
limit2
```

# Завдання 2

```python
import sympy as sp
from IPython.display import display

x = sp.symbols("x")

# ---------- Function 1 ----------
f1 = sp.cot( x**(1/sp.log(x)) )

f1_prime = sp.diff(f1, x)
f1_third = sp.diff(f1, x, 3)

display(f1_prime)
display(f1_third)
```

```python
import sympy as sp
from IPython.display import display

a, b = sp.symbols("a b")

# ---------- Function 2 ----------
f2 = ((a**x + b**x)/2)**(1/x)

f2_prime = sp.diff(f2, x)
f2_third = sp.diff(f2, x, 3)

display(f2_prime)
display(f2_third)
```

# Завдання 3

```python
import sympy as sp

x = sp.symbols("x")

# ---------- First integral ----------
expr1 = (x * sp.cos(x)) / sp.sin(x)**3
indef_integral = sp.integrate(expr1, x)
print("Невизначений інтеграл:")
display(indef_integral)
```

```python
import sympy as sp

x = sp.symbols("x")

# ---------- Second integral ----------
expr2 = x**2 * sp.exp(3*x)
def_integral = sp.integrate(expr2, (x, 0, 1))
print("\nЗначення визначеного інтегралу:")
display(def_integral)
print(f"Числове значення: {def_integral.evalf()}")
```
