```python
import math

def y(x, a, k, c):
    term_1 = (math.log(x) + a**2)**(1/3) / (0.47 * x**2)
    term_2 = abs(0.47 * x**2 - (10**4 / 7) * math.cos(k)**2)
    term_3 = c / x
    return term_1 - term_2 - term_3
```

```python
x = 0.29
a = -2.4
k = 3
c = 1.52
```

```python
result = y(x, a, k, c)
print(f"y = {result:.4e}")
```
